```python
def check_entry(student_name, student_list):
    if student_name in student_list:
        print("Entry Allowed")
    else:
        print("Entry Not Allowed")

registered_students = [
    "aman", "siya", "avantee", "rashmin", "sneha"
]

name = input("Enter student name: ")

check_entry(name, registered_students)
```
