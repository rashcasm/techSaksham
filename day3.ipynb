```python
# make a code to login using username and password and send friend request to a friend
users = {
    "aman": "aman",
    "siya": "siya",
    "avantee": "avantee",
    "rashmin": "rashmin",
    "sneha": "sneha"
}
def login(username, password):
    if username in users and users[username] == password:
        print("Login Successful")
        return True
    else:
        print("Login Failed")
        return False
def send_friend_request(sender, receiver):
    if sender in users and receiver in users:
        print(f"Friend request sent from {sender} to {receiver}")
    else:
        print("Sender or receiver does not exist")
username = input("Enter username: ")
password = input("Enter password: ")
if login(username, password):
    friend_name = input("Enter the name of the friend you want to send a request to: ")
    send_friend_request(username, friend_name) 
```
