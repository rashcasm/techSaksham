```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset
data = pd.read_csv('environmental_socioeconomic.csv')
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
# Normalize the features using StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
import scipy.cluster.hierarchy as sch
link = sch.linkage(data_scaled, method = 'complete')
dend = sch.dendrogram(link)
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters = 2, linkage = 'ward', metric = 'euclidean')
labels = model.fit_predict(data_scaled)
labels
```

```python
import matplotlib.pyplot as plt
plt.scatter(data_scaled[:,0], data_scaled[:,1], c = model.labels_, cmap = 'rainbow')
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
type(cancer)
cancer
```

```python
#Creating the dataframe from this bunch data
cancer_df = pd.DataFrame(cancer['data'], columns = cancer['feature_names'])
cancer_df.head()
```

```python
cancer_df.info()
```

```python
from sklearn.preprocessing import StandardScaler
# Normalize the features using StandardScaler
scaler = StandardScaler()
cancer_df_scaled = scaler.fit_transform(cancer_df)
cancer_df_scaled[:5]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
X_new = pca.fit_transform(cancer_df_scaled)
X_new.shape
```

```python
plt.scatter(X_new[:,0], X_new[:,1], c = cancer['target'], cmap = 'plasma')
plt.show()
```
