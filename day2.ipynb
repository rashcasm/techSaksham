```python
account1 = {
    "name": "Rashmin",
    "balance": 2000
}
account2 = {
    "name": "Chaudhari",
    "balance": 500
}
accounts = [account1, account2]
expense = 500
for account in accounts:
    if account["balance"] >= expense:
        account["balance"] -= expense
        print(f"{account['name']} paid {expense}. Remaining balance: {account['balance']}")
    else:
        print(f"{account['name']} does not have enough balance to pay {expense}.")
```

```python
account1 = {"balance": 5000}
transaction_limit = 2
transactions = 0
for i in range(5):
    if transactions >= transaction_limit:
        print("Daily transaction limit is reached.")
        break
    if account1["balance"] >= 1000:
        account1["balance"] -= 1000
        transactions += 1
        print(f"Transaction {transactions}: Paid 1000. Remaining balance: {account1['balance']}")
    else:
        print("Insufficient balance.")
        break
```
